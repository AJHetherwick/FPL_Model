```python
import os
import pickle

def get_match_report(report_url):
  FBREF_BASE_URL = 'https://fbref.com'
  report_url = FBREF_BASE_URL + report_url
  report = pd.read_html(report_url)
  #todo cleanup the report
  return report

error_matches = {}
matches = {}
if os.path.exists("match-report-data.pkl"):
  with open("match-report-data.pkl", "rb") as f:
      matches = pickle.load(f)
```

```python
#there are 380 games in the season
REQUESTS_PER_MINUTE = 2
print(f'Expected time to complete: {380 / REQUESTS_PER_MINUTE} minutes or {(380 / REQUESTS_PER_MINUTE) / 60} hours')
```

```python
for i, fixture in fixtures.iterrows():
  match_name = f'Week-{fixture.Wk[0]} {fixture.Home[0]} vs {fixture.Away[0]}'
  try:
    if match_name not in matches:
      matches[match_name] = (
          get_match_report(fixture['Match Report'][1])
      )

      with open("match-report-data.pkl", "wb") as f:
        pickle.dump(matches, f)

      print(f'Processed {match_name}')
    else:
      print(f'Already processed {match_name}')
  except Exception as e:
    print(f'Error processing {match_name}')
    error_matches[match_name] = str(e)
    print(e)
    print(f'Sleeping for 1 hour at time:{time.asctime()} to overcome likely block restart if incorrect')
    time.sleep(60*60)
  #sleep to avoid getting ip blocked
  time.sleep(60 / REQUESTS_PER_MINUTE)
print(*[f'{m} : {len(d)}' for m,d in matches.items()], sep='\n')
```

```python
import pandas as pd
from functools import reduce

MERGE_COL = '#'
def safe_merge(left, right):
    # Drop overlapping columns from right (except 'Player')
    left = left.dropna(subset=[MERGE_COL])
    right = right.dropna(subset=[MERGE_COL])
    overlap = set(left.columns) & set(right.columns) - {MERGE_COL}
    right = right.drop(columns=list(overlap))
    return pd.merge(left, right, on=MERGE_COL, how='outer')

pretty_beautiful_joined_matches = []
for j, (match, match_dfs) in enumerate(matches.items()):
    home_df = None
    away_df = None

    selected_dfs = []
    for i, df in enumerate(match_dfs[3:10]):
        df = df.copy(deep=True)
        if isinstance(df.columns, pd.MultiIndex):
            #print(df.columns)
            df.columns = [str(col[-1]) if col[0].lower().startswith('unnamed') else '_'.join(map(str, col)) for col in df.columns.values]
            #print(df.columns)
        if MERGE_COL in df.columns:
            selected_dfs.append(df)
    home_df = reduce(safe_merge, selected_dfs)
    home_df.loc[:, 'match-name'] = match
    home_df['#'] = home_df['#'].astype(int)

    selected_dfs = []
    for i, df in enumerate(match_dfs[10:17]):
        df = df.copy(deep=True)
        if isinstance(df.columns, pd.MultiIndex):
            #print(df.columns)
            df.columns = [str(col[-1]) if col[0].lower().startswith('unnamed') else '_'.join(map(str, col)) for col in df.columns.values]
            #print(df.columns)
        if MERGE_COL in df.columns:
            selected_dfs.append(df)
    away_df = reduce(safe_merge, selected_dfs)
    away_df.loc[:, 'match-name'] = match
    away_df['#'] = away_df['#'].astype(int)

    match_joined_df = pd.concat([home_df, away_df])
    pretty_beautiful_joined_matches.append(match_joined_df)

# pretty_beautiful_joined_matches contains a dataframe for each match with all the stats for each player in the match in each row except overall posession and formation stats and stats for each shot taken in the match
pretty_beautiful_joined_matches[1]
```

```python
import pandas as pd

# FPL_CSV_URL = "https://raw.githubusercontent.com/vaastav/Fantasy-Premier-League/master/data/2024-25/gws/merged_gw.csv"
#Note: Reading from this url directly will mess you up
#they merged 2 parts of the season the first has 42 columns, the second has 49 columns (tracking 7 columns of manager stats)
#The csv header has 42 columns, it is a bomb
#I split it into 2 csvs manually and load them below

ALL_FIXTURES_URL = 'https://raw.githubusercontent.com/vaastav/Fantasy-Premier-League/refs/heads/master/data/2024-25/fixtures.csv'
ALL_TEAMS_URL = 'https://raw.githubusercontent.com/vaastav/Fantasy-Premier-League/refs/heads/master/data/2024-25/teams.csv'

fpl_df_a = pd.read_csv('vaastav-gw-data-merged-1.csv')
fpl_df_b = pd.read_csv('vaastav-gw-data-merged-2.csv')
fpl_df_b = fpl_df_b.drop(columns=fpl_df_b.columns[fpl_df_b.columns.str.startswith('mng_')])
fpl_df = pd.concat([fpl_df_a, fpl_df_b])

teams = pd.read_csv(ALL_TEAMS_URL)
fixture = pd.read_csv(ALL_FIXTURES_URL)
fixture["team_h"] = fixture["team_h"].map(teams.set_index("id")["name"])
fixture["team_a"] = fixture["team_a"].map(teams.set_index("id")["name"])
team_map = {
    "Ipswich": "Ipswich Town",
    "Leicester": "Leicester City",
    "Man City": "Manchester City",
    "Man Utd": "Manchester Utd",
    "Newcastle": "Newcastle Utd",
    "Nott'm Forest": "Nott'ham Forest",
    "Spurs": "Tottenham"
}
fixture["team_h"] = fixture["team_h"].replace(team_map)
fixture["team_a"] = fixture["team_a"].replace(team_map)

fixture = fixture.add_prefix('fixture_')

# Merge
fpl_df = fpl_df.merge(fixture, left_on='fixture', right_on='fixture_id', how='left', indicator=True)

print(fpl_df._merge.value_counts())
print(fpl_df[fpl_df._merge == 'left_only'].fixture.unique())
fpl_df = fpl_df.drop(columns=['_merge'])

fpl_df['match-name'] = fpl_df.apply(lambda x: f'{x.fixture_team_h} vs {x.fixture_team_a}', axis=1)
# Display the first few rows of the DataFrame
fpl_df
```
